```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
#Task 3 (Exploratory Data Analysis)
print("Average measurements per species:")
print(df.groupby('species').mean()) # Summarize findings

print("\nData Correlation:")
print(df.select_dtypes(include='number').corr()) # Identify trends
```
